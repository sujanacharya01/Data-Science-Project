```python
df
```

```python
df.drop(['S. No.'], axis=1, inplace=True)  # Here i have droppes serial Number column
```

```python
df.head()
```

```python
# Here i created a new 'Total Cases' column to know 'Total covid Cases'
df['Total Cases'] = df['Total Confirmed cases (Indian National)'] + df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
# To know Over all total cases
print("Total number of cases found in india is : ", df['Total Cases'].sum())
```

```python
# 'Active Cases' column has been created to know active cases. 

print(" ")
print("Kindly go through the 'Active Cases' column to know active cases by state wise")

df['Active Cases'] = df['Total Cases']-(df['Cured']+df['Death'])
df
```

```python
print("Total number of active cases in india is : ",df['Active Cases'].sum())
```

```python
df.style.background_gradient(cmap='Blues')
```

```python
# To know 'total active cases' by 'state' wise

print(" Kindly go through below data to know Active Cases by State wise")
Total_Active_Cases = df.groupby('Name of State / UT')['Active Cases'].sum().sort_values(ascending=False).to_frame()
Total_Active_Cases.style.background_gradient(cmap='Blues')
```

```python
confirmed=df.groupby('Date').sum()['Confirmed'].reset_index()
death=df.groupby('Date').sum()['Deaths'].reset_index()
rec=df.groupby('Date').sum()['Recovered'].reset_index()
```
